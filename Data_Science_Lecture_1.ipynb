```python
a=np.array([[1,2,3],[5,6,7]])
k=a.flatten()
b=np.reshape(a,(6,1))
print(b)
c=np.transpose(b)
print(c)
print(k)
```

```python
np.random.seed(4)  #### fix the random value in the some patterns
a=np.random.random_sample((5,2))*5
print(a)
b=np.min(a)
c=np.argmin(a)  #### give the location of the minimum value
print(b)
print(c)
x=np.mean(a)
y=np.var(a)
z=np.std(a)
print(x)
print(y)
print(z)
```

```python
a=np.random.random_sample((3,3))
print(a)
b=np.exp(a)
print(b)
```

```python
a=np.random.random_sample((3,3))
b=np.linalg.det(a)  #### linalg is the class where all the linear algebra operator is defined you can explore the other function in this class
print(b)
c=np.linalg.matrix_rank(a)
print(c)
```
