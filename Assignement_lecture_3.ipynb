```python
# Loading the packages
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import linregress
```

```python
url_train = "https://raw.githubusercontent.com/shala2020/shala2020.github.io/master/Lecture_Materials/Assignments/DataScience/L2/train.csv"
train = pd.read_csv(url_train)
train.head(10)
```

```python
url_movies = "https://raw.githubusercontent.com/shala2020/shala2020.github.io/master/Lecture_Materials/Google_Colab_Notebooks/DataScience/L2/moviesData.csv"
movies = pd.read_csv(url_movies)
movies.head(10)
```

```python
movies.info()
```

```python
movies.isna().sum()
```

```python
correlation_matrix=movies.corr()
correlation_matrix
```

```python
plt.figure(figsize=(8,8))
sns.heatmap(correlation_matrix,annot=True,cmap='coolwarm', fmt=".2f", linewidths=.5)
plt.show()
```

```python
mask = np.triu(np.ones_like(correlation_matrix, dtype=bool))
plt.figure(figsize=(8,8))
sns.heatmap(correlation_matrix,annot=True,cmap='coolwarm', fmt=".2f", linewidths=.5,mask=mask)
plt.show()
```

```python
sns.regplot(x=movies['audience_score'],y=movies['imdb_rating'],line_kws={"color":"blue"}, scatter_kws={"color":"red","alpha":0.3,"s":100})
slope,intercept,r_value, p_value, std_err=linregress(movies['audience_score'],movies['imdb_rating'])
equation = f'y = {slope:.2f}x + {intercept:.2f}'
equation
print(r_value,p_value,std_err)
```

```python
df = sns.load_dataset('iris')
sns.scatterplot(x="petal_length",y="petal_width",data=df,hue='species',s=100)
plt.legend(loc='lower right')
```

```python
df.head()
```

```python
from pylab import *

plt.figure(figsize=(10,10))
subplot(2,2,1)
sns.kdeplot(data=df['sepal_length'],shade=True,label="sepal_length",color='r')

subplot(2,2,2)
sns.kdeplot(data=df['sepal_width'],shade=True,label="sepal_width",color='y')

subplot(2,2,3)
sns.kdeplot(data=df['petal_length'],shade=True,label="petal_length",color='g')

subplot(2,2,4)
sns.kdeplot(data=df['petal_width'],shade=True,label="petal_width",color='b')
plt.show()
```

```python
plt.figure(figsize=(10,10))
subplot(2,2,1)
sns.kdeplot(data=df['sepal_length'],shade=True,label="sepal_length",color='r',cumulative=True)

subplot(2,2,2)
sns.kdeplot(data=df['sepal_width'],shade=True,label="sepal_width",color='y',cumulative=True)

subplot(2,2,3)
sns.kdeplot(data=df['petal_length'],shade=True,label="petal_length",color='g',cumulative=True)

subplot(2,2,4)
sns.kdeplot(data=df['petal_width'],shade=True,label="petal_width",color='b',cumulative=True)
plt.show()
```
